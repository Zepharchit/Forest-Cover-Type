Finding correlation between data


```python
# calculate the correlation matrix
df_corr = df.iloc[:,0:11]
corr = df_corr.corr()
plt.figure(figsize=(10,8))
# plot the heatmap
sns.heatmap(corr, 
        xticklabels=corr.columns,

            yticklabels=corr.columns,annot=True,cmap="BuPu")
```

# PLOTTING among various correlated features(+ve or -ve)

Plotting Correlated features,setting a threshold of 0.47

```python
corr_list = []
cols=df_corr.columns 
size=10
threshold = 0.50
#Search for the highly correlated pairs
for i in range(0,size): #for 'size' features
    for j in range(i+1,size): #avoid repetition
        if (corr.iloc[i,j] >= threshold and corr.iloc[i,j] < 1) or (corr.iloc[i,j] < 0 and df_corr.iloc[i,j] <= -threshold):
            corr_list.append([df.iloc[i,j],i,j]) #store correlation and columns index

#Sort to show higher ones first            
s_corr_list = sorted(corr_list,key=lambda x: -abs(x[0]))
print('Correlated Features')
for v,i,j in s_corr_list:
    print ("%s and %s" %(cols[i],cols[j]))
```

```python
for v,i,j in s_corr_list:
    sns.pairplot(df, hue="Cover_Type", height=6, x_vars=cols[i],y_vars=cols[j] )
```

DATA PREPARATION

```python
# Training Data
df.head()
y_training = df.Cover_Type.values
df_train = df.drop(['Id','Soil_Type7','Soil_Type15','Cover_Type'],axis=1)
df_train.head()
size = 11
X_train_temp = df_train.iloc[:,:size]

#Test Data
df_test_new = df_test.drop(['Id','Soil_Type7','Soil_Type15'],axis=1)
df_test_new.head()
X_test_new  = df_test_new.iloc[:,:size]
```

```python
# Taking non categorical variable for scaling 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_temptr = sc.fit_transform(X_train_temp)
X_tempte = sc.fit_transform(X_test_new)
```

```python
df_train.head()
```

```python
#creating the new dataset with scaled values
rows,col = df_train.shape
row_t,col_t= df_test_new.shape
print(df_train.shape)
print(df_test_new.shape)
```

```python
X_TRain = np.concatenate((X_temptr,df_train.iloc[:,10:col]),axis=1)
X_TEst = np.concatenate((X_tempte,df_test_new.iloc[:,10:col]),axis=1)
```

```python
X_TRain.shape
X_TEst.shape
```

```python
# Creating Training and testing values from training dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X_TRain,y_training,test_size=0.2,random_state=0)
```

```python
#Deploying the following Algorithms
#SVM,Log Regression, Random Forest,Extra Tree Classifier,KNN
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Logistic Regression
lg = LogisticRegression()
lg.fit(x_train,y_train)
lg_predict = lg.predict(x_test)
acc_lg = lg.score(x_test,y_test)
print(acc_lg)
```

```python
# Random Forest
rf = RandomForestClassifier(n_estimators=400,class_weight="balanced",criterion="entropy",random_state=0)
rf.fit(x_train,y_train)
rf_predict = rf.predict(x_test)
rf_acc = rf.score(x_test,y_test)
print(rf_acc)
```

```python
cov_type = rf.predict(X_TEst)
print(cov_type)
```

```python
#SVM
sv = svm.SVC(kernel='linear',class_weight="balanced",random_state=0)
sv.fit(x_train,y_train)
sv_predict = sv.predict(x_test)
sv_acc = sv.score(x_test,y_test)
print(sv_acc)
```

```python
# Extra Tree Classifier
etc = ExtraTreesClassifier(n_estimators=300,class_weight="balanced",random_state=0)
etc.fit(x_train,y_train)
etc_predict=etc.predict(x_test)
etc_acc = etc.score(x_test,y_test)
print(etc_acc)
```

```python
cov_type_etc = etc.predict(X_TEst)
print(cov_type_etc)
```

```python
# KNN
knn = KNeighborsClassifier(n_neighbors=5,algorithm='ball_tree')
knn.fit(x_train,y_train)
knn_predit = knn.predict(x_test)
knn_acc = knn.score(x_test,y_test)
print(knn_acc)
```

```python
#Gradient Boosting
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier(learning_rate=0.65 , n_estimators= 250 ,max_depth = 9)
clf.fit(x_train,y_train)
clf_predict = clf.predict(x_test)
clf_acc = clf.score(x_test,y_test)
print(clf_acc)
```
